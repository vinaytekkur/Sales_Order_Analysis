```python
# 2- write a program to get city wise count of return orders
```

```python
df_orders = pd.read_csv('orders.txt')
df_returns = pd.read_csv('returns.txt')
```

```python
# merge data sets
df_orders_returns = pd.merge(left=df_orders, right=df_returns, how='left', left_on='order_id', right_on='order_id')
```

```python
df_orders_returns['return_reason'].fillna(value='Not_Returned', inplace=True)
```

```python
# 3- write a program to print cities where we have all 3 kinds of returns (others,bad quality,wrong items)
```

```python
order_df = pd.read_csv('orders.txt')
return_df = pd.read_csv('returns.txt')
```

```python
# Merge datasets
order_return_df = pd.merge(left=order_df, right=return_df, how='left', left_on='order_id', right_on='order_id')
```

```python
order_return_df['return_reason'].fillna(value='Not_Returned', inplace=True)
```

```python
fc = order_return_df['return_reason'] != 'Not_Returned'
city_nonreturns_orders = order_return_df[fc].groupby('city', as_index=False)['return_reason'].nunique()
```

```python
city_nonreturns_orders.columns = ['City', 'Non_returned_orders_count']
```

```python
city_nonreturns_orders[city_nonreturns_orders['Non_returned_orders_count'] >= 3]
```

```python
# 4- write a program to find cities where not even a single order was returned.
```

```python
df_o = pd.read_csv('orders.txt')
df_r = pd.read_csv('returns.txt')
```

```python
get_order_return = pd.merge(left=df_o, right=df_r, how='left', left_on='order_id', right_on='order_id')
```

```python
get_order_return['return_reason'].fillna(value='Not_Returned', inplace=True)
```

```python
fc = get_order_return['return_reason'] != 'Not_Returned'
```

```python
orders_returned_city = list(get_order_return[fc]['city'].unique())
```

```python
orders_not_returned_city = list(get_order_return[~get_order_return['city'].isin(orders_returned_city)]['city'].unique())
```

```python
# City name which does not returned single orders
orders_not_returned_city
```

```python
# 5- write a program to find top 3 cities by sales
df_order = pd.read_csv('orders.txt')
df_returns = pd.read_csv('returns.txt')
```
